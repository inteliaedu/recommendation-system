# Correción en atípicos

## Importar Librerías

```python
import pandas as pd
import gc
import pyarrow
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

## Lectura de datos

```python
df = pd.read_parquet(engine="pyarrow", path="/content/drive/My Drive/UCM/tfm_code/data/data_modelo_train.parquet")
```

## Transformación de datos

```python
# Detectando Z-score Outliers en acumulado de cajas
df['CF'] = df['CF'].astype(float)
upper_limit = df['CF'].mean() + 3*df['CF'].std()
df['CF_Z_OUTLIERS'] = np.where(df['CF']>upper_limit,1,0)
print("Cuenta última compra outliers Z-score (CF): "+str(df['CF_Z_OUTLIERS'].sum()))
print("Total registros última compra (CF): "+str(df['CF_Z_OUTLIERS'].count()))
print("Porcentaje de Outliers por Z-score Cuenta: "+str(100* df['CF_Z_OUTLIERS'].sum()/df['CF_Z_OUTLIERS'].count()))
print("Porcentaje de Outliers por Z-score Suma (CF): "+str(100* df.loc[df['CF_Z_OUTLIERS'] == 1]['CF'].sum()/df['CF'].sum()))
print("Upper Limit (CF): "+ str(upper_limit))
```

```python
# Detectando Z-score Outliers en última compra
df['LAST_CF'] = df['LAST_CF'].astype(float)
upper_limit = df['LAST_CF'].mean() + 3*df['LAST_CF'].std()
df['LAST_CF_OUTLIERS'] = np.where(df['LAST_CF']>upper_limit,1,0)
print("Cuenta última compra outliers Z-score (CF): "+str(df['LAST_CF_OUTLIERS'].sum()))
print("Total registros última compra (CF): "+str(df['LAST_CF_OUTLIERS'].count()))
print("Porcentaje de Outliers por Z-score Cuenta: "+str(100* df['LAST_CF_OUTLIERS'].sum()/df['LAST_CF_OUTLIERS'].count()))
print("Porcentaje de Outliers por Z-score Suma (CF): "+str(100* df.loc[df['LAST_CF_OUTLIERS'] == 1]['LAST_CF'].sum()/df['LAST_CF'].sum()))
print("Upper Limit (CF): "+ str(upper_limit))
```

```python
# Removiendo Outliers
df = df.loc[(df['CF_Z_OUTLIERS'] == 0) & (df['LAST_CF_OUTLIERS'] == 0)]
```

```python
# Removiendo clientes con menos de 4 compras a la semana
filtered = df.groupby(['CLIENTE_ID','SEMANA']).filter(lambda x: len(x) >= 5)
```

```python
# Removiendo posibles registros con compras negativas o en 0
filtered = filtered[filtered['LAST_CF']>0]
filtered = filtered[filtered['CF']>0]
```

```python
# Creación de nuevas variables
filtered['ROTATION_MEAN_CF']= filtered['LAST_CF']/filtered['MEAN_CF']
filtered['ROTATION_MEAN_DAYS']= filtered['RECENCY']/filtered['MEAN_DIFF_DAYS']
filtered['ROTATION_MEDIAN_DAYS']= filtered['RECENCY']/filtered['MEDIAN_DIFF_DAYS']
filtered['CF_LOG']= np.log(filtered['CF'])
```

```python
# Removiendo columnas
filtered.drop(columns=['LAST_CF','MEAN_CF','MEDIAN_CF','MEAN_DIFF_DAYS','MEDIAN_DIFF_DAYS','CF','CF_Z_OUTLIERS','LAST_CF_OUTLIERS'],inplace=True)
```

```python
# Normalización de columnas con escalado Min Max
cols_to_norm = ['FRECUENCIA','ROTATION_MEAN_CF', 'ROTATION_MEDIAN_DAYS', 'ROTATION_MEAN_DAYS','RECENCY', 'DURATION_CLIENT', 'TAMANIO','UNIDAD_EMPAQUE']
filtered[cols_to_norm] = MinMaxScaler().fit_transform(filtered[cols_to_norm])
```

```python
# Guardando datos
filtered.to_parquet("/content/drive/My Drive/UCM/tfm_code/data/data_modelo_train_cleaned.parquet", engine='pyarrow', compression='snappy')
```

## Transformación de datos en set de prueba

```python
df = pd.read_parquet(engine="pyarrow", path="/content/drive/My Drive/UCM/tfm_code/data/data_modelo_test.parquet")
```

```python
df = df[df['LAST_CF']>0]
df = df[df['CF']>0]
```

```python
df['ROTATION_MEAN_CF']= df['LAST_CF']/df['MEAN_CF']
df['ROTATION_MEAN_DAYS']= df['RECENCY']/df['MEAN_DIFF_DAYS']
df['ROTATION_MEDIAN_DAYS']= df['RECENCY']/df['MEDIAN_DIFF_DAYS']
df['CF_LOG']= np.log(df['CF'])
```

```python
# Density Plot and Histogram of sum CF
sns.set_theme()
sns.set(font_scale=1.5)
plt.figure(figsize=(15,8));
ax = sns.distplot(df['CF'], hist=True, kde=True, 
             bins=int(200/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});
ax.set(xlabel='Cajas Físicas')
ax.set(ylabel='Densidad')
ax.set_title("Cajas Físicas acumuladas en últimas 7 semanas", fontsize=20);
```

```python
# Density Plot and Histogram of sum CF
sns.set_theme()
sns.set(font_scale=1.5)
plt.figure(figsize=(15,8));
ax = sns.distplot(df['CF_LOG'], hist=True, kde=True, 
             bins=int(200/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4});
ax.set(xlabel='Logaritmo Cajas Físicas')
ax.set(ylabel='Densidad')
ax.set_title("Logaritmo Cajas Físicas acumuladas en últimas 7 semanas", fontsize=20);
```

```python
df.drop(columns=['LAST_CF','MEAN_CF','MEDIAN_CF','MEAN_DIFF_DAYS','MEDIAN_DIFF_DAYS','CF'],inplace=True)
```

```python
cols_to_norm = ['FRECUENCIA','ROTATION_MEAN_CF', 'ROTATION_MEDIAN_DAYS', 'ROTATION_MEAN_DAYS','RECENCY', 'DURATION_CLIENT', 'TAMANIO','UNIDAD_EMPAQUE']
df[cols_to_norm] = MinMaxScaler().fit_transform(df[cols_to_norm])
```
