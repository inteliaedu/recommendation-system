# Importando librerías

```python
import pandas as pd
import pyarrow
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Desarrollo de matriz de correlación

```python
df = pd.read_parquet(engine="pyarrow", path="/content/drive/My Drive/UCM/tfm_code/data/data_modelo_train_cleaned.parquet")
```

```python
df.drop(columns=['CLIENTE_ID','PRODUCTO_ID','ANIO','SEMANA'], inplace=True)
```

```python
df.columns
```

```python
cols_to_correlate = ['FRECUENCIA', 'RECENCIA', 'DURACION_DEL_CLIENTE',  'ROTACION_MEDIA_CF', 'ROTACION_MEDIA_DIAS','ROTACION_MEDIANA_DIAS', 'CF_LOG', 'TAMANIO','UNIDAD_EMPAQUE']
```

```python
dict = {'RECENCY': 'RECENCIA',
        'DURATION_CLIENT': 'DURACION_DEL_CLIENTE',
        'ROTATION_MEAN_CF': 'ROTACION_MEDIA_CF',
        'ROTATION_MEAN_DAYS': 'ROTACION_MEDIA_DIAS',
        'ROTATION_MEDIAN_DAYS': 'ROTACION_MEDIANA_DIAS',
        }
 
# call rename () method
df.rename(columns=dict,
          inplace=True)
```

```python
#correlation matrix
cmap = sns.diverging_palette(220, 10, as_cmap=True)
corrmat= df[cols_to_correlate].corr()
plt.figure(figsize=(12,12))  
sns.heatmap(corrmat,annot=True,  cmap=cmap,center=0);
```
